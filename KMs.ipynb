```python
#引入函數庫
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
#產生都是兩個數字的100組資料
X = np.random.rand(100,2)
```

```python
X[0]
```

```python
#顯示
plt.scatter(X[:,0], X[:,1], s=50)
```

```python
#匯入KMeans
from sklearn.cluster import KMeans
```

```python
#讓他分3類
clf = KMeans(n_clusters=3)
```

```python
#開始學習
clf.fit(X)
```

```python
#取得結果
clf.labels_
```

```python
#顯示結果
plt.scatter(X[:,0], X[:,1], c=clf.labels_)
```
